```python
# 2) Calcula el saldo final
saldo = 0

for m in movimientos:
    if m["tipo"] == "abono":
        saldo += m["monto"]
    elif m["tipo"] == "cargo":
        saldo -= m["monto"]

print("Saldo final:", saldo)

print("\n" + "="*50)
```

## 8. Tu tarea: completar la implementación

Se debe implementar los siguientes métodos:

- `obtener(posicion)`
- `eliminar(posicion)`
- `mostrar_todos()`
- `calcular_saldo()`

### Recomendaciones
- validar si la posición existe
- decidir cuándo conviene usar `return`
- probar cada método apenas se complete

```python
class ListaMovimientos:
    def __init__(self):
        self.movimientos = []

    def esta_vacia(self):
        return len(self.movimientos) == 0

    def agregar(self, movimiento):
        self.movimientos.append(movimiento)

    def tamano(self):
        return len(self.movimientos)

    def obtener(self, posicion):
        if 0 <= posicion < len(self.movimientos):
            return self.movimientos[posicion]
        else:
            print("Error: Posición inválida.")
            return None

    def eliminar(self, posicion):
        if 0 <= posicion < len(self.movimientos):
            return self.movimientos.pop(posicion)
        else:
            print("Error: Posición inválida, no se pudo eliminar.")
            return None

    def mostrar_todos(self):
        if self.esta_vacia():
            print("La lista de movimientos está vacía.")
        else:
            print("--- Movimientos ---")
            for i, movimiento in enumerate(self.movimientos):
                print(f"{i}. {movimiento['descripcion']} | {movimiento['monto']} | {movimiento['tipo']}")
            print("-------------------")

    def calcular_saldo(self):
        saldo = 0
        for movimiento in self.movimientos:
            if movimiento["tipo"] == "abono":
                saldo += movimiento["monto"]
            elif movimiento["tipo"] == "cargo":
                saldo -= movimiento["monto"]
        return saldo
```
